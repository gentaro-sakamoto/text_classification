```python
from gensim import corpora
import pandas as pd
import numpy as np

dict = corpora.Dictionary.load_from_text('./comments.dict')
data = [(k, v, dict.dfs[k]) for k, v in dict.iteritems()]

df = pd.DataFrame(data, columns=['id', 'word', 'freq'])
df.sort_values(by='freq', ascending=False)
```
