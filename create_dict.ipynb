```python
import MeCab
from IPython import embed
from IPython.terminal.embed import InteractiveShellEmbed
import os
from gensim import corpora
import re
import simplejson
```

```python
def validate(node):
  pos = node.feature.split(",")[0]
  if pos != u'名詞':
    return False

  if node.posid in exclude_pos_ids:
    return False

  if re.search('[あ-んア-ン]', node.surface):
    return False

  if len(node.surface) == 1 and re.search('[一-龥]', node.surface):
    return False

  if re.search('[0-9０-９]', node.surface):
    return False

  return True
```

```python
encoding = "utf-8"

mecab = MeCab.Tagger ('-d /usr/local/lib/mecab/dic/mecab-ipadic-neologd')
text_dir_path = './texts/comments'

effective_words = []
for filename in sorted(os.listdir(text_dir_path)):
  if filename.endswith('.txt'):
    effective_words_by_text = []
    print('reading {}'.format(filename) )
    with open(text_dir_path + '/' + filename, encoding=encoding) as file:
      text = file.read()
    node = mecab.parseToNode(text)

    exclude_pos_ids = [
      38, # 名詞,一般 ex. 1つ
      41, # 名詞,固有名詞,一般 ex. -1
      42, # 名詞,固有名詞
      43, # 名詞,固有名詞
      44, # 名詞,固有名詞
      45, # 名詞,固有名詞,組織
      46, # 名詞,固有名詞,地域,一般 ex. 10区、14号
      47, # 名詞,固有名詞
      48, # 名詞,数
      53, # 名詞,接尾,助数詞
      63, # 名詞,非自立
      64, # 名詞,非自立
      65, # 名詞,非自立
      66, # 名詞,非自立

    ]

    debug_targets = ['以上']

    while node:
        if validate(node) is True:
          if node.surface in debug_targets:
            embed()
          effective_words_by_text.append(node.surface)
        node = node.next
    effective_words.append(effective_words_by_text)

print('total {} texts'.format(len(effective_words)))
with open('./effective_words.json', 'w') as f:
    simplejson.dump(effective_words, f)

dictionary = corpora.Dictionary(effective_words)
dictionary.filter_extremes(no_below=5, no_above=0.3)
dictionary.save_as_text('./comments.dict')
```
